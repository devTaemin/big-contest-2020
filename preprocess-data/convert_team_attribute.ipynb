```python
def convert_team_csv(filepath):
    file = pd.read_csv(filepath)
    file_columns = file.columns.to_list()
    new_index = ["G_ID", "GDAY_DS", "T_ID", "VS_T_ID", "HEADER_NO", "TB_SC",
             "WHIP", "LOB", "ERA", "FIP", "H_9", "K_9", "BB_9", "HR_9", "K_BB",
             "oAVG", "oSLG", "oOBP", "P_WHIP_RT", "P2_WHIP_RT", "CB_WHIP_RT",
             "AVG", "OBP", "SLG", "OPS", "RC", "GPA", "BA", "SECA", "TA", "XR", "ISO",
             "wOBA", "P_HRA_RT", "DER", "PF", "H_A", "WLS"]
    
    #출처 statiz 
    PF_dict = {"SS":1062, "LT":1041, "NC":1035, "HH":1025, "KT":1020, "HT":1004, "WO":995, "SK":971, "LG":939, "OB":939}
    
    df = pd.DataFrame(columns = new_index)

    for i in range(len(file)):
        row = file[i:i+1]     #1 줄씩 불러오기 -> 값 추출
                              # -> 추출한 값으로 세이버매트릭스 변환
                              # -> 한 줄에 dataFrame에 입력후 append해서 write
        
        # 앞에 p가 prefix인 것은 pitcher, h는 hitter
        G_ID = row["G_ID"].values[0]
        GDAY_DS = row["GDAY_DS"].values[0]
        T_ID = row["T_ID"].values[0]
        VS_T_ID = row["VS_T_ID"].values[0]
        HEADER_NO = row["HEADER_NO"].values[0]
        TB_SC = row["TB_SC"].values[0]
        p_CG_CK = row["p_CG_CK"].values[0]
        p_WLS = row["p_WLS"].values[0]
        p_HOLD = row["p_HOLD"].values[0]
        p_INN2 = row["p_INN2"].values[0]
        p_BF = row["p_BF"].values[0]
        p_PA = row["p_PA"].values[0]
        p_AB = row["p_AB"].values[0]
        p_HIT = row["p_HIT"].values[0]
        p_H2 = row["p_H2"].values[0]
        p_H3 = row["p_H3"].values[0]
        p_HR = row["p_HR"].values[0]
        p_SB = row["p_SB"].values[0]
        p_CS = row["p_CS"].values[0]
        p_SH = row["p_SH"].values[0]
        p_SF = row["p_SF"].values[0]
        p_BB = row["p_BB"].values[0]
        p_IB = row["p_IB"].values[0]
        p_HP = row["p_HP"].values[0]
        p_KK = row["p_KK"].values[0]
        p_GD = row["p_GD"].values[0]
        p_WP = row["p_WP"].values[0]
        p_BK = row["p_BK"].values[0]
        p_ERR = row["p_ERR"].values[0]
        p_R = row["p_R"].values[0]
        p_ER = row["p_ER"].values[0]
        p_P_WHIP_RT = row["p_P_WHIP_RT"].values[0]
        p_P2_WHIP_RT = row["p_P2_WHIP_RT"].values[0]
        p_CB_WHIP_RT = row["p_CB_WHIP_RT"].values[0]
        h_PA = row["h_PA"].values[0]
        h_AB = row["h_AB"].values[0]
        h_RBI = row["h_RBI"].values[0]
        h_RUN = row["h_RUN"].values[0]
        h_HIT = row["h_HIT"].values[0]
        h_H2 = row["h_H2"].values[0]
        h_H3 = row["h_H3"].values[0]
        h_HR = row["h_HR"].values[0]
        h_SB = row["h_SB"].values[0]
        h_CS = row["h_CS"].values[0]
        h_SH = row["h_SH"].values[0]
        h_SF = row["h_SF"].values[0]
        h_BB = row["h_BB"].values[0]
        h_IB = row["h_IB"].values[0]
        h_HP = row["h_HP"].values[0]
        h_KK = row["h_KK"].values[0]
        h_GD = row["h_GD"].values[0]
        h_ERR = row["h_ERR"].values[0]
        h_LOB = row["h_LOB"].values[0]
        h_P_HRA_RT = row["h_P_HRA_RT"].values[0]
        h_P_AB_CN = row["h_P_AB_CN"].values[0]
        h_P_HIT_CN = row["h_P_HIT_CN"].values[0]

        # 변환 값
        # --- TEMP ----------------------------------------------
        p_H1 = (p_HIT - p_H2 - p_H3 - p_HR)
        h_H1 = (h_HIT - h_H2 - h_H3 - h_HR)
        RC_A = (h_HIT + h_BB - h_CS + h_HP - h_GD) #출루능력
        RC_B = 0.24 * (h_BB - h_IB + h_HP) + (0.62 * h_SB) + 0.5 * (h_SH + h_SF) - 0.03 * h_KK #진루능력
        RC_C = (h_AB + h_BB + h_HP + h_SH + h_SF) #주어진 기회
        #G_ID = G_ID.values[0]
        #GDAY_DS = GDAY_DS.values[0]
        #T_ID = T_ID.values[0]
        #VS_T_ID = VS_T_ID.values[0]
        #HEADER_NO = HEADER_NO.values[0]
        #TB_SC = TB_SC.values[0]
        #p_WLS = p_WLS.values[0]
        #p_P_WHIP_RT = p_P_WHIP_RT.values[0]
        #p_P2_WHIP_RT = p_P2_WHIP_RT.values[0]
        #p_CB_WHIP_RT = p_CB_WHIP_RT.values[0]
        
        
        # --- PITCHER -------------------------------------------
        WHIP = round((p_HIT + p_BB) / (p_INN2 / 3), 3)
        LOB = round((p_HIT + p_BB + p_HP - p_R) / (p_HIT + p_BB + p_HP - (1.4 * p_HR)), 3)
        ERA = round(((9 * p_ER) / (p_INN2 / 3)), 3)
        FIP = round((13 * p_HR) + (3 * (p_BB + p_HP - p_IB)) - ((2 * p_KK) / (p_INN2 / 3)), 3)
        H_9 = round((9 * p_HIT) / (p_INN2 / 3), 3)
        K_9 = round((9 * p_KK) / (p_INN2 / 3), 3)
        BB_9 = round((9 * p_BB) / (p_INN2 / 3), 3)
        HR_9 = round((9 * p_HR) / (p_INN2 / 3), 3)
        K_BB = round((p_KK/p_BB), 3)
        oAVG = round((p_HIT / (p_PA - p_BB - p_HP - p_SF - p_SB)), 3)
        oSLG = round((p_H1 + (2 * p_H2) + (3 * p_H3) + (4 * p_HR)) / (p_PA - p_BB - p_HP - p_SF - p_SB), 3)
        oOBP = round((p_HIT + p_BB) / (p_AB + p_BB + p_HP + p_SF))
        P_WHIP_RT = round(p_P_WHIP_RT, 3)
        P2_WHIP_RT = round(p_P2_WHIP_RT, 3)
        CB_WHIP_RT = round(p_CB_WHIP_RT, 3)
        # --- HITTER -------------------------------------------
        AVG = round((h_HIT / h_AB), 3)
        OBP = round((h_HIT + h_BB + h_HP) / (h_AB + h_BB + h_HP + h_SF), 3)
        SLG = round((h_H1 + h_H2 * 2 + h_H3 * 3 + h_HR * 4) / (h_AB), 3)
        OPS = (OBP + SLG)
        RC = round(((2.4 * RC_C + RC_A) * (3 * RC_C + RC_B)) / (9 * RC_C) - (0.9 * RC_C), 3)
        GPA = round(((1.8 * OBP + SLG) / 4), 3)
        BA = round((h_HIT / h_AB), 3)
        SECA = round((((h_H2 * 2) + (h_H3 * 3) + (h_HR * 4) + h_BB + (h_SB - h_CS)) / h_AB), 3)
        TA = round(((h_H1 + (h_H2 * 2) + (h_H3 * 3) + (h_HR * 4) + h_HP + h_BB + h_SB) / (h_AB - h_HIT + h_CS + h_GD)), 3)
        XR = round(((0.5 * h_H1) + (0.72 * h_H2) + (1.04 * h_H3) + (1.44 * h_HR) + 0.34 * (h_BB + h_HP - h_IB) + (0.25 * h_IB) +
              (0.18 * h_SB) - (0.32 * h_CS) - 0.09 * (h_AB - h_HIT - h_KK) - (0.098 * h_KK) - (0.37 * h_GD) +(0.04 * h_SH)), 3)
        ISO = (SLG - AVG)
        wOBA = round(((0.7 * (h_BB - h_IB) + (0.73 * h_HP) + (0.89 * h_H1) + (1.27 * h_H2) + (1.61 * h_H3) + (2.07 * h_HR) + (0.25 * h_SB) -
                (0.5 * h_CS)) / (h_AB - h_IB)), 3)
        P_HRA_RT = round((h_P_HRA_RT), 3)
        DER = round(((p_PA - p_HIT - p_KK - p_BB - p_HP - p_IB - p_BK) / (p_PA - p_HR - p_KK - p_BB - p_HP)) , 3)
        
        if (TB_SC == 'T'):
            PF = PF_dict[VS_T_ID]
            H_A = 0 #AWAY
        else:
            PF = PF_dict[T_ID]
            H_A = 1 #HOME
        
        if (p_WLS == 'W'):
            WLS = 1 #WIN
        else:
            WLS = 0 #DRAW, LOSE
        # ------------------------------------------------------
        
        new_data = [(G_ID, GDAY_DS, T_ID, VS_T_ID, HEADER_NO, TB_SC, WHIP, LOB, ERA, FIP, H_9, K_9, BB_9, HR_9, K_BB,oAVG, oSLG, oOBP, P_WHIP_RT, P2_WHIP_RT, 
                   CB_WHIP_RT, AVG, OBP, SLG, OPS, RC, GPA, BA, SECA, TA, XR, ISO, wOBA, P_HRA_RT, DER, PF, H_A, WLS)]
        
        new_row = pd.DataFrame(new_data, columns = new_index)
            
        df = df.append(new_row, ignore_index = True)
        
    fname = "convert_" + filepath.split('.')[0].split('\\')[1] 
    fpath = filepath.split('.')[0].split('\\')[0][:4] + "_convert"
    if ((os.path.isdir(os.getcwd() + "\\" + fpath) == False)):
        os.makedirs('./'+fpath)
    df.to_csv(fpath + "\\" + fname + ".csv")
            
          
```
